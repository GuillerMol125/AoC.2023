## SAMPLE INPUT PROCESSING

```python
input= ["seeds: 79 14 55 13",
"",
"seed-to-soil map:",
"50 98 2",
"52 50 48",
"",
"soil-to-fertilizer map:",
"0 15 37",
"37 52 2",
"39 0 15",
"",
"fertilizer-to-water map:",
"49 53 8",
"0 11 42",
"42 0 7",
"57 7 4",
"",
"water-to-light map:",
"88 18 7",
"18 25 70",
"",
"light-to-temperature map:",
"45 77 23",
"81 45 19",
"68 64 13",
"",
"temperature-to-humidity map:",
"0 69 1",
"1 0 69",
"",
"humidity-to-location map:",
"60 56 37",
"56 93 4"]

begin_tag = 'seed'
final_tag = 'location'
```

```python
index = 0
line = input[index]
mapping_dict = {}
while not 'seeds' in line:
    index += 1
    line = input[index]

tag, seeds = line.split(':')
seed_list = seeds.split()

while index < len(input):
    while not 'map' in line:
        index += 1
        if index == len(input):
            break
        line = input[index]
    print(line)
    mapname, maptag = line.split()

    map_list = []
    index += 1
    line = input[index]
    while line:
        this_map = line.split()
        map_list.append(this_map)
        index += 1
        if index == len(input):
            break
        line = input[index]

    mapping_dict[mapname] = map_list
    
```

```python
def map_value(value, map_list):
    for mapping in map_list:
        offset = value - int(mapping[1])
        if  (offset >=0) and (offset < int(mapping[2])):
            mapped = int(mapping[0]) + offset
            return mapped
    return value
```

```python
def find_next_mapping(current_tag, map_dict):
    for map_tag in map_dict:
        if current_tag + '-to-' in map_tag:
            return map_tag
    return None
```

```python
# DO ALL THE MAPPINGS AND GET MINIMUM LOCATION VALUE FOR THE SEEDS
location_list = []
for initial_seed_range in seed_list:
    current_tag = begin_tag
    current_value = int(initial_seed_range)
    while not current_tag == final_tag:
        map_tag = find_next_mapping(current_tag, mapping_dict)
        #print(f'Apply {map_tag} to {current_value}')
        next_value = map_value(current_value, mapping_dict[map_tag])
        next_tag = map_tag.split('-to-')[1]
        print(f'Apply {map_tag} to {current_value} maps into {next_value}')
        current_value = next_value
        current_tag = next_tag
    location_list.append(current_value)
print(location_list)
print(min(location_list))
```

```python
input= ["seeds: 79 14 55 13",
"",
"seed-to-soil map:",
"50 98 2",
"52 50 48",
"",
"soil-to-fertilizer map:",
"0 15 37",
"37 52 2",
"39 0 15",
"",
"fertilizer-to-water map:",
"49 53 8",
"0 11 42",
"42 0 7",
"57 7 4",
"",
"water-to-light map:",
"88 18 7",
"18 25 70",
"",
"light-to-temperature map:",
"45 77 23",
"81 45 19",
"68 64 13",
"",
"temperature-to-humidity map:",
"0 69 1",
"1 0 69",
"",
"humidity-to-location map:",
"60 56 37",
"56 93 4"]

begin_tag = 'seed'
final_tag = 'location'
```

```python
index = 0
line = input[index]
mapping_dict = {}
while not 'seeds' in line:
    index += 1
    line = input[index]

tag, seeds = line.split(':')
seed_list = seeds.split()

while index < len(input):
    while not 'map' in line:
        index += 1
        if index == len(input):
            break
        line = input[index]
    print(line)
    mapname, maptag = line.split()

    map_list = []
    index += 1
    line = input[index]
    while line:
        this_map = line.split()
        map_list.append(this_map)
        index += 1
        if index == len(input):
            break
        line = input[index]

    mapping_dict[mapname] = map_list
    
```

```python
def overlapping_range(range_orig, range_dest):
    range_orig_init = int(range_orig[0])
    range_orig_end = range_orig_init + int(range_orig[1]) - 1

    range_dest_init = int(range_dest[0])
    range_dest_end = range_dest_init + int(range_dest[1]) - 1 

    common_init = max(range_orig_init, range_dest_init)
    common_end = min(range_orig_end, range_dest_end)
    
    if common_end < common_init:
        overlap = None 
        remaining = [range_orig]
    else:
        overlap = [common_init, common_end - common_init + 1]
        remaining = []
        if range_orig_init < common_init:
            remaining.append([range_orig_init, common_init - range_orig_init])
        if range_orig_end > common_end:
            remaining.append([common_end + 1, range_orig_end - common_end])
    return overlap, remaining

def map_range_to_mapping(range, mapping):
    range_dest = mapping[1:]
    offset_mapping = int(mapping[0]) - int(mapping[1])
    overlap, remaining_list = overlapping_range(range, range_dest)
    if overlap:
        overlap[0] = overlap[0] + offset_mapping
    return overlap, remaining_list

def map_range_to_list(range_list, map_list):
    mapped_list = []
    remaining_list = range_list
    for mapping in map_list:
        next_remaining_list = []
        for remaining_range in remaining_list:
            mapped, remaining = map_range_to_mapping(remaining_range, mapping)
            if mapped:
                mapped_list.append(mapped)
            next_remaining_list = next_remaining_list + remaining
        remaining_list = next_remaining_list
    
    mapped_list = mapped_list + remaining_list
    
    return mapped_list

def get_seed_ranges(seeds):
    seed_ranges = []
    seed_range = []
    for seed_value in seeds:
        seed_range.append(int(seed_value))
        if len(seed_range) == 2:
            seed_ranges.append(seed_range)
            seed_range = []
    return seed_ranges
```

```python
# DO ALL THE MAPPINGS AND GET MINIMUM LOCATION VALUE FOR THE SEED RANGES
location_list = []
seed_range_list = get_seed_ranges(seed_list)
current_tag = begin_tag
current_range_list = seed_range_list
    
while not current_tag == final_tag:
    map_tag = find_next_mapping(current_tag, mapping_dict)
    #print(f'Apply {map_tag} to {current_value}')
    next_range_list = map_range_to_list(current_range_list, mapping_dict[map_tag])
    next_tag = map_tag.split('-to-')[1]
    print(f'Apply {map_tag} to {current_range_list} maps into {next_range_list}')
    current_range_list = next_range_list
    current_tag = next_tag

location_list = current_range_list

print(location_list)
print(min(location_list))
```
