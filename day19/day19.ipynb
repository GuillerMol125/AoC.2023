```python
def wokflow_process(piece, workflow_dict, workflow_name):
    workflow = workflow_dict[workflow_name]
    for step in workflow[:-1]:
        check, dest = step.split(':')
        idx = check[0]
        op = check[1]
        val = check[2:]
        expr = piece[idx] + op + val
        if eval(expr):
            return dest
    return workflow[-1]
```

```python
def full_process(piece):
    workflow_name = 'in'

    while not(workflow_name=='A') and not(workflow_name=='R'):
        print(f'Process piece {piece} thru wf {workflow_name} {workflow_dict[workflow_name]}')
        workflow_name = wokflow_process(piece,workflow_dict,workflow_name)
        print(workflow_name)
    return workflow_name
```

```python
accepted_list = []
for piece in pieces:
    final_place = full_process(piece)
    if final_place == 'A':
        accepted_list.append(piece)
print(accepted_list)
sumfinal = 0
for piece in accepted_list:
    sumfinal += sum([int(value) for value in piece.values()])
print(sumfinal)
```

```python
def wokflow_range_process(ranges, workflow_dict, workflow_name):
    workflow = workflow_dict[workflow_name]
    result_dict = {}

    for step in workflow[:-1]:
        check, dest = step.split(':')
        idx = check[0]
        op = check[1]
        val = int(check[2:])
        range = ranges[idx]

        if op == '>':
            if min(range) > val: # FULL PASS
                new_ranges = ranges.copy()
                result_dict[dest] = new_ranges
                return result_dict
            elif max(range) > val: # PARTIAL PASS
                new_ranges = ranges.copy()
                new_ranges[idx] = [val+1, max(range)]
                ranges[idx] = [min(range), val]
                result_dict[dest] = new_ranges
            else:               # FULL FAIL
                continue

        if op == '<':
            if max(range) < val: # FULL PASS
                new_ranges = ranges.copy()
                result_dict[dest] = new_ranges
                return result_dict
            elif min(range) < val: # PARTIAL PASS
                new_ranges = ranges.copy()
                new_ranges[idx] = [min(range), val - 1]
                ranges[idx] = [val, max(range)]
                result_dict[dest] = new_ranges
            else:               # FULL FAIL
                continue

    new_ranges = ranges.copy()
    result_dict[workflow[-1]] =  new_ranges

    return result_dict
```
