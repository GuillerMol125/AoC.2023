## PROCESS FIRST SAMPLE

```python
input = ["0 3 6 9 12 15",
"1 3 6 10 15 21",
"10 13 16 21 30 45"]
```

```python
def get_increments(sequence):
    return sequence[1:] - sequence[0:-1]

def get_next_value(sequence):
    if (max(sequence) == 0) and (min(sequence) == 0):
        return 0
    else:
        return sequence[-1] + get_next_value(get_increments(sequence))
```

```python
# PROCESS ALL
sum_next_values = 0
for line in input:
    sequence = np.array([int(x) for x in line.split()])
    increments = get_increments(sequence)
    next_value = get_next_value(sequence)
    sum_next_values += next_value
    print(f'Sequence {sequence} has increments {increments} and next value {next_value}')
sum_next_values
```
