## SAMPLE INPUT PROCESSING

```python
import pandas as pd
import numpy as np
import os, sys
```

```python
sample_input = ["467..114..",
"...*......",
"..35..633.",
"......#...",
"617*......",
".....+.58.",
"..592.....",
"......755.",
"...$.*....",
".664.598.."]
```

```python
def get_numbers(line):
    number_list = []
    char_index = 0
    top_len = len(line)-1
    line = line + '.'
    while char_index <= top_len:
        begin = char_index
        end = char_index-1
        number = 0
        while line[char_index] >= '0' and line[char_index] <= '9' and char_index <= top_len:
            end = char_index
            number = 10*number + int(line[char_index])
            char_index += 1
        if end >= begin:
            number_list.append((number,begin,end))
        while (line[char_index] < '0' or line[char_index] > '9') and char_index <= top_len:
            char_index += 1
    return number_list
```

```python
# GENERATE INPUT DATA
number_list = []
for line in sample_input:
    number_list.append(get_numbers(line))

print(number_list)

np_sample_input = []
for line in sample_input:
    np_sample_input.append([x for x in line])

np_sample_input = np.array(np_sample_input)
```

```python
def find_symbol(nparray, row, begin, end):
    max_row, max_col = nparray.shape

    begin_row = max(0, row-1)
    end_row = min(max_row-1, row + 1)

    begin_col = max(0, begin - 1)
    end_col = min(max_col - 1, end + 1)

    #print(f'Checking within rows {begin_row} and {end_row} and cols {begin_col} and {end_col}')
    for check_row in range(begin_row, end_row+1):
        for check_col in range(begin_col, end_col+1):
            char_check = nparray[check_row, check_col]
            if (char_check < '0' or char_check > '9') and not(char_check == '.'):
                #print(f'Fount at row {check_row} and col {check_col} value {char_check}')
                return True

    return False
```

```python
valid_numbers = []
row_ix = 0
for number_list_row in number_list:
    for number in number_list_row:
        #print(number)
        if find_symbol(np_sample_input, row_ix,number[1], number[2]):
            valid_numbers.append(number[0])
    row_ix += 1
```

## SECOND SAMPLE

```python
sample_input = ["467..114..",
"...*......",
"..35..633.",
"......#...",
"617*......",
".....+.58.",
"..592.....",
"......755.",
"...$.*....",
".664.598.."]
```

```python
# GENERATE INPUT DATA
number_list = []
for line in sample_input:
    number_list.append(get_numbers(line))

print(number_list)

np_input = []
for line in sample_input:
    np_input.append([x for x in line])

np_input = np.array(np_input)
```

```python
def fill_gears(nparray, row, begin, end, number):
    max_row, max_col = nparray.shape

    begin_row = max(0, row-1)
    end_row = min(max_row-1, row + 1)

    begin_col = max(0, begin - 1)
    end_col = min(max_col - 1, end + 1)

    #print(f'Checking within rows {begin_row} and {end_row} and cols {begin_col} and {end_col}')
    gear_list = []
    for check_row in range(begin_row, end_row+1):
        for check_col in range(begin_col, end_col+1):            
            if nparray[check_row, check_col] == '*':
                gear_list.append((check_row, check_col, number))
            #print(f'filling up {nparray[check_row, check_col]}')
    return gear_list
```

```python
row_ix = 0
all_gears = {}
for number_list_row in number_list:
    for number in number_list_row:
        print(number)
        new_gears = fill_gears(np_input, row_ix,number[1], number[2], number[0])
        print(new_gears)
        for gear in new_gears:
            try:
                all_gears[str(gear[0]) + '_' + str(gear[1])].append(gear[2])
            except KeyError:
                all_gears[str(gear[0]) + '_' + str(gear[1])] = [gear[2]]
    row_ix += 1
print(all_gears)
```

```python
# GET THE VALUES FOR GEAR 2
total_value = 0
for idx, list in all_gears.items():
    if len(list) == 2:
        gear_value = list[0]*list[1]
        print(gear_value)
        total_value += gear_value

total_value
```
