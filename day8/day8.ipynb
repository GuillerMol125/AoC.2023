```python
# BUILD TRAVEL PATH LOOPS
current_position_list = starting_node_list
moves = 0
finished_target = len(current_position_list)
print(f'Target is: {finished_target}')
finished = 0
loop_moves_dict = {}
for starting_node in starting_node_list:
    print(f'Checking starting node {starting_node}')
    finished_moves = []
    visited_nodes = {}
    visited_nodes[starting_node] = 0
    moves = 0
    current_node = starting_node
    last_visit_same_node = 0
    while True:
        loop_value = None
        instruction_position = 0
        for instruction in instructions:
            instruction_position += 1
            moves += 1
            current_node = map_dict[current_node][instruction]
            current_node_signature = current_node + str(instruction_position)
            if current_node[-1] == 'Z':
                print(f'Found ending {current_node} at move {moves}')
                finished_moves.append(moves)
            try:
                last_visit_same_node = visited_nodes[current_node_signature]
                loop_value = [last_visit_same_node, moves]
                print(f'Found loop between {last_visit_same_node} and {moves}')
                break
            except KeyError:
                visited_nodes[current_node_signature] = moves
        if loop_value:
            loop_moves_dict[starting_node] = {}
            loop_moves_dict[starting_node]['FINISHES'] = finished_moves
            loop_moves_dict[starting_node]['LOOP'] = loop_value
            break
```

```python
# PRINT LOOPS
finish_list = []
loop_sizes = []
for starting, content in loop_moves_dict.items():
    loopsize = content['LOOP'][1] - content['LOOP'][0]
    print(f'starting node {starting} has a loop size of {loopsize}')
    finish_list.append(content['FINISHES'][0])
    loop_sizes.append(loopsize)
finish_list, loop_sizes
```

```python
def aligned(positions):
    for position in positions:
        if not position == positions[0]:
            return False
    return True
```

```python
# GET DIVISORS FOR EACH
prime_factors = []
for current_finish in finish_list:
    print(f'Getting prime factors for {current_finish}')
    divisor_list = []
    divisor = 2
    while current_finish > 1:
        if current_finish % divisor == 0:
            divisor_list.append(divisor)
            current_finish = current_finish / divisor
            divisor = 2
        else:
            divisor += 1
        if divisor*divisor > current_finish:
            divisor_list.append(current_finish)
            current_finish = 1
    prime_factors.append(divisor_list)
```
