```python
input = ["#.##..##.",
"..#.##.#.",
"##......#",
"##......#",
"..#.##.#.",
"..##..##.",
"#.#.##.#.",
"",
"#...##..#",
"#....#..#",
"..##..###",
"#####.##.",
"#####.##.",
"..##..###",
"#....#..#"]
```

```python
mapping = {'#':1, '.':0}
terrain_list = []
terrain = []
for line in input:
    row = [mapping[x] for x in line]
    if len(row) == 0:
        terrain_list.append(np.array(terrain))
        terrain = []
    else:
        terrain.append(row)
        row = []
if len(terrain) > 0:
    terrain_list.append(np.array(terrain))
```

```python
def check_vertical_reflection(array, column):
    num_rows, num_cols = array.shape
    overlapping_cols = min(column+1, num_cols-column-1)
    
    if overlapping_cols <= column:
        array_left = array[:,column:column-overlapping_cols:-1]
    else:
        array_left = array[:,column::-1]
    array_right = array[:,column+1:column+overlapping_cols+1]
    reflection = np.sum(np.abs(array_right - array_left))
    
    return reflection

def check_horizontal_reflection(array, row):
    num_rows, num_cols = array.shape
    overlapping_rows = min(row+1, num_rows-row-1)
    
    if overlapping_rows <= row:
        array_up = array[row:row-overlapping_rows:-1,:]
    else:
        array_up = array[row::-1,:]
    array_down = array[row+1:row+overlapping_rows+1,:]
    reflection = np.sum(np.abs(array_up - array_down))
    
    return reflection
```

```python
check_vertical_reflection(terrain_list[0],0)
```

```python
result = 0
for terrain in terrain_list:
    terrain_rows, terrain_cols = terrain.shape
    for col in range(terrain_cols-1):
        reflection = check_vertical_reflection(terrain,col)
        print(f'Vertical reflection at col {col} is {reflection}')
        result += (reflection==0)*(col+1)
    for row in range(terrain_rows-1):
        reflection = check_horizontal_reflection(terrain,row)
        print(f'Horizontal reflection at row {row} is {reflection}')
        result += 100*(reflection==0)*(row+1)
result
```
