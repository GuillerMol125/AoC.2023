## PROCESS FIRST SAMPLE

```python
input = '''R 6 (#70c710)
D 5 (#0dc571)
L 2 (#5713f0)
D 2 (#d2c081)
R 2 (#59c680)
D 2 (#411b91)
L 5 (#8ceee2)
U 2 (#caa173)
L 1 (#1b58a2)
U 2 (#caa171)
R 2 (#7807d2)
U 3 (#a77fa3)
L 2 (#015232)
U 2 (#7a21e3)'''
```

```python
input = input.split('\n')
```

```python
direction_map = {'R':(0,1), 'L':(0,-1), 'U':(-1,0), 'D':(1,0)}

crossing_map = {('.','.','.','.'): [],
                ('.','.','.','#'): [],
                ('.','.','#','.'): [], 
                ('.','.','#','#'): ['UD'],
                ('.','#','.','.'): [],
                ('.','#','.','#'): ['RD'],
                ('.','#','#','.'): ['RU'], 
                ('.','#','#','#'): ['RUD'], 
                ('#','.','.','.'): [], 
                ('#','.','.','#'): ['LD'], 
                ('#','.','#','.'): ['LU'], 
                ('#','.','#','#'): ['LUD'], 
                ('#','#','.','.'): ['LR'],
                ('#','#','.','#'): ['LRD'], 
                ('#','#','#','.'): ['LRU'], 
                ('#','#','#','#'): ['LRUD']
                 } # LEFT, RIGHT, UP, DOWN
```

```python
position = (0,0)
map = set()
map.add(position)

for line in input:
    direction, length, color = line.split()
    dx,dy = direction_map[direction]
    for step in range(int(length)):
        x, y = position
        next_position = (x+dx,y+dy)
        map.add(next_position)
        position = next_position
print(map)
```

```python
# GET THE PERIMETER
min_row, max_row = min(map,key=lambda x: x[0])[0], max(map,key=lambda x: x[0])[0]
min_col, max_col = min(map,key=lambda x: x[1])[1], max(map,key=lambda x: x[1])[1]
map_grid = []
perimeter = 0
for row in range(min_row, max_row+1):
    row_grid = []
    for col in range(min_col, max_col+1):
        if (row,col) in map:
            row_grid.append('#')
            perimeter += 1
        else:
            row_grid.append('.')
    map_grid.append(row_grid)
    print(''.join(row_grid))
print(perimeter)
```

```python
# FILL THE INTERIOR
min_row, max_row = min(map,key=lambda x: x[0])[0], max(map,key=lambda x: x[0])[0]
min_col, max_col = min(map,key=lambda x: x[1])[1], max(map,key=lambda x: x[1])[1]

total_cave = 0
for row in range(min_row, max_row+1):
    inside = False
    pattern_list = []
    for col in range(min_col, max_col+1):
        if (row,col) in map:    # CASE FOR #
            total_cave += 1
            pattern = []
            for direction_value, (drow, dcol) in direction_map.items():
                if (row+drow,col+dcol) in map:
                    pattern.append(direction_value)
            pattern_list.append(pattern)
        else:                   # CASE FOR .
            if len(pattern_list) > 0:
                first_pattern = pattern_list[0]
                last_pattern = pattern_list[-1]
                if ('D' in first_pattern and 'U' in last_pattern) or ('U' in first_pattern and 'D' in last_pattern):
                    inside = not(inside)
                pattern_list = []
            total_cave += inside

total_cave
```

## PROCESS SECOND SAMPLE

```python
input = '''R 6 (#70c710)
D 5 (#0dc571)
L 2 (#5713f0)
D 2 (#d2c081)
R 2 (#59c680)
D 2 (#411b91)
L 5 (#8ceee2)
U 2 (#caa173)
L 1 (#1b58a2)
U 2 (#caa171)
R 2 (#7807d2)
U 3 (#a77fa3)
L 2 (#015232)
U 2 (#7a21e3)'''

input = input.split('\n')
```

```python
commands = []
direction_decode = {'0':'R','1':'D', '2':'L', '3':'U'}
points = {}
position = (0,0)
x,y = position
for line in input:
    _,_,command = line.split(' ')
    command = command[1:-1]

    direction = direction_decode[command[-1]]
    length = int(command[1:-1], 16)

    dx, dy = direction_map[direction]

    x, y = x + dx*length, y + dy*length
    if x in points.keys():
        points[x].append(y)
        points[x].sort()
    else:
        points[x] = [y]

points
    
```

```python
# COMPUTE ALL AREA:
segments_length = 0
previous_row = 0
total_area = 0
current_point_list = []
for row in sorted(list(points.keys())):
    point_list = points[row]
    print('-----------------------------------------------------')
    print(f'New point row {row}')
    print(f'New points {point_list}')
    
    block_points = segments_length*(row - previous_row - 1) 
    total_area += block_points
    previous_row = row

    print(f'Block points {block_points}')
    
    current_point_list.sort()
    point_list.sort()

    # Add all segments in this exact row
    begin_list = []
    end_list = []
    for i in range(0,len(current_point_list),2):
        begin = current_point_list[i]
        end = current_point_list[i+1]
        begin_list.append(begin)
        end_list.append(end)
    for i in range(0,len(point_list),2):
        begin = point_list[i]
        end = point_list[i+1]
        begin_list.append(begin)
        end_list.append(end)
    begin_list.sort()
    end_list.sort()

    all_points = begin_list + end_list
    
    all_points = set(all_points)
    all_points = list(all_points)
    all_points.sort()
    coverage_level = 0
    coverage_so_far = 0
    previous_point = 0
    print(f'List of begginnings {begin_list}')
    print(f'List of ends {end_list}')
    for checkpoint in all_points:
        print(f'Checking point {checkpoint} with coverage level {coverage_level}')
        if coverage_level:
            coverage_so_far += checkpoint - previous_point
            print(f'\tAdding extra row coverage {checkpoint - previous_point} at point {checkpoint} (total so far {coverage_so_far})')
        coverage_level += begin_list.count(checkpoint)
        coverage_level -= end_list.count(checkpoint)
        if coverage_level == 0: #EVERY SEGMENT END ADDS 1
            print(f'\tAdding 1 more at point {checkpoint} (total so far {coverage_so_far})')
            coverage_so_far += 1
        previous_point = checkpoint
    print(f'Row coverage {coverage_so_far}')
    total_area += coverage_so_far*1

    current_point_list += point_list
    current_point_list.sort()

    new_point_list = [x for x in current_point_list if current_point_list.count(x)==1]
    current_point_list = new_point_list
    print(f'Cumulative points {current_point_list}')

    segments_length = 0
    for i in range(0,len(current_point_list),2):
        begin = current_point_list[i]
        end = current_point_list[i+1] + 1
        segments_length += end-begin
    print(f'Total Area: {total_area}')

total_area
```
